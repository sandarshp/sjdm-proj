```python
import hssm
import hssm.plotting
import numpy as np
import pandas as pd
import arviz as az
import seaborn as sns
from matplotlib import pyplot as plt
import bambi as bmb
import statsmodels.api as sm
from pymer4.models import Lmer
from scipy import stats

```

```python
pairings_df = pd.read_csv('pairings_demo.csv')
```

```python
pairings_df.head()
```

```python
pairings_df['phq_scale'] = pairings_df['phq_score']/27
```

```python
# define conditions
conditions = [
    (pairings_df["valence_left"].eq("positive") & pairings_df["valence_right"].eq("negative")),  # left pos, right neg
    (pairings_df["valence_left"].eq("negative") & pairings_df["valence_right"].eq("positive"))   # left neg, right pos
]

# define choices
choices = [-1, 1]

# assign values
pairings_df["delta_valence"] = np.select(conditions, choices, default=0)


pairings_df.head()
```

```python
model_vars = ["r_diff_norm", "delta_valence",
              "phq_scale", "age_z", "female_binary", "subject"]

pairings_df_clean = pairings_df.dropna(subset=model_vars)
```

```python
model_demo2 = hssm.HSSM(
    model = 'levy',
    data = pairings_df_clean,
    include = [
        {
            "name": "v",
            "formula": "v ~ 1 + r_diff_norm + female_binary + delta_valence*phq_scale*age_z + (1 + r_diff_norm + delta_valence|subject)",       
        }
               
    ],
    
)

ddm_samples_demo2 = model_demo2.sample(sampler = "mcmc", tune = 2000, draws = 10000, chains = 4, cores = 1,target_accept=0.95)
ddm_samples_demo2.to_netcdf('./models/ddm_model_demo2.nc')
print("Over")
```
